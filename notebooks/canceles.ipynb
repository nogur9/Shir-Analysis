```python
import pandas as pd
import seaborn as sns
import plotly
import matplotlib.pyplot as plt
import datetime
```

```python
df = pd.read_csv("../canceled.csv")
can_df = pd.read_excel("../data/Stripe Quit.xlsx")

email = 'Customer Email'
name = 'Customer Name'

df["email_clean"] = df[email].str.lower().str.strip()
df["clean_name"] = df[name].str.lower().str.strip()
```

```python
email_1 = can_df['Email 1 Home'].str.lower().str.strip()
email_2 = can_df['Email 2 Work'].str.lower().str.strip()
can_df["clean_name"] = can_df["Full Name"].str.lower().str.strip()
can_df["concat_name"] = can_df["First name"].str.lower().str.strip() + " " + can_df["Last name"].str.lower().str.strip()
```

```python
miss_in_can = df[~(df["email_clean"].isin(email_1) | df["email_clean"].isin(email_2))]
miss_in_df = can_df[~(email_1.isin(df["email_clean"]) | email_2.isin(df["email_clean"]))]
```

```python
miss_in_can = miss_in_can[~ miss_in_can["clean_name"].isin(can_df["clean_name"])]
miss_in_can = miss_in_can[~ miss_in_can["clean_name"].isin(can_df["concat_name"])]
```

```python
miss_in_df = miss_in_df[~ miss_in_df["clean_name"].isin(df["clean_name"])]
miss_in_df = miss_in_df[~ miss_in_df["concat_name"].isin(df["clean_name"])]
```

```python
miss_in_df.to_csv("missing in Stripe.csv", index = 0)
```

```python
miss_in_can.to_csv("missing in Shirs.csv", index = 0)
```
